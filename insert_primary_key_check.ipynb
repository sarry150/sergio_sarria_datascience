```python
def check_repeated_line(dataframe, newLine, **kwargs):
    try:
        params = {}
        headers = dataframe.columns.tolist()
        repeated = {}

        for key, value in kwargs.items():
            params[key] = value

        for key, value in params:
            if key in headers:
                for i in range(dataframe.shape[0]):
                    if value == dataframe[key][i]:
                        repeated[key] = value
                        break
            else:
                return print('The parameter ' + param + 'does not exist in the dataframe')
        if params == repeated:
            return print('The line you want to introduce does already exists')
        else:
            dataframe.append(newLine)
        return dataframe
    except Exception as error:
        print(str(error))
                
                    
```
